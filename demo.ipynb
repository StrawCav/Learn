```python
# Randomized primality test
import random
import math

def randomized_test(n, trials=10):
    '''
    trials: number of random I  [2, floor(sqrt(n))] to test
    '''
    if n < 2:
        return False
    if n in (2, 3):
        return True

    upper = math.isqrt(n)
    for _ in range(trials):
        I = random.randint(2, upper)
        if n % I == 0:
            return False 
    return True

nums = [2, 3, 4, 5, 9, 16, 17, 25, 29, 30]
for n in nums:
    print(n, randomized_test(n))
```
